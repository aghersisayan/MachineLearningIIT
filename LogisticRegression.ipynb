```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Cambia la ruta según la ubicación de tu archivo en Google Drive
ruta_archivo = '/content/drive/MyDrive/AMZN_data_1999_2022.csv'

# Cargar el archivo en un DataFrame de pandas
df = pd.read_csv(ruta_archivo)

# Ahora puedes trabajar con tu DataFrame df
```

```python
plt.figure(figsize = (18,9))
plt.plot(range(df.shape[0]),(df['Open']+df['Close'])/2.0)
plt.xticks(range(0,df.shape[0],500),df['Date'].loc[::500],rotation=45)
plt.xlabel('Date',fontsize=18)
plt.ylabel('Mid Price',fontsize=18)
plt.show()
```

```python
# Sort DataFrame by date
df = df.sort_values('Date')
# Imprimir los primeros registros del DataFrame
print(df.head())
```

```python
# Crear la nueva columna "Average"
df['Average'] = (df['Open'] + df['Close']) / 2

# Imprimir los primeros registros del DataFrame con la nueva columna
print(df.head())
```

```python
print(df.tail())
```

```python
# Calcular la media móvil de las últimas 200 filas y crear la nueva columna "MA200"
df['MA200'] = df['Average'].rolling(window=200, min_periods=1).mean()

# Imprimir los primeros registros del DataFrame con la nueva columna
print(df.head())
```

```python
print(df.tail())
```

```python
# Calcular la media móvil de las últimas 50 filas y crear la nueva columna "MA50"
df['MA50'] = df['Average'].rolling(window=50, min_periods=1).mean()

# Imprimir los primeros registros del DataFrame con la nueva columna
print(df.head())
```

```python
# Assuming your date column is named 'Date'
df['Datetime'] = pd.to_datetime(df['Date'])  # Convert the 'Date' column to datetime format
reference_date = pd.to_datetime('1999-01-03')
# Calculate the number of days since the reference date
df['Day'] = (df['Datetime'] - reference_date).dt.days

print(df.head())
```

```python

df = df.drop(['Date','Datetime'], axis=1)
print(df.head())
```

```python
# Crear la nueva columna de clasificación
df['Clase'] = 0  # Establecer inicialmente todos los valores como 'Down'
df.loc[df['Average'] >= df['Average'].shift(1), 'Clase'] = 1

# Imprimir los primeros registros del DataFrame con la nueva columna
print(df.head())
```

```python
# Borrar las columnas "Volume" y "Adj Close"
df = df.drop(['Volume', 'Adj Close'], axis=1)

# Imprimir los primeros registros del DataFrame después de eliminar las columnas
print(df.head())
```

```python
plt.figure(figsize = (18,9))
plt.plot(range(df.shape[0]),(df['Low']+df['High'])/2.0)
plt.xticks(range(0,df.shape[0],500),df['Day'].loc[::500],rotation=45)
plt.xlabel('Day',fontsize=18)
plt.ylabel('Mid Price',fontsize=18)
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Seleccionar solo las columnas numéricas para la normalización
columns_to_normalize = ['Day', 'Open', 'Close', 'High', 'Low', 'Average', 'MA200', 'MA50']


import matplotlib.pyplot as plt
import seaborn as sns

# Establecer el estilo para los gráficos (opcional)
sns.set(style="whitegrid")

# Gráfico antes de la normalización
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[columns_to_normalize])
plt.title('Diagrama de Caja de Columnas Seleccionadas Antes de la Normalización')
plt.show()

# Normalizar las columnas seleccionadas
df[columns_to_normalize] = tf.keras.utils.normalize(df[columns_to_normalize])

# Gráfico después de la normalización
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[columns_to_normalize])
plt.title('Diagrama de Caja de Columnas Seleccionadas Después de la Normalización')
plt.show()
```

```python

# Obtener el número total de filas
total_rows = df.shape[0]

# Definir los porcentajes para la división
train_percentage = 0.6
validation_test_percentage = 0.2

# Calcular el número de filas para cada conjunto
train_size = int(total_rows * train_percentage)
validation_test_size = int(total_rows * validation_test_percentage)

# Puedes usar iloc para seleccionar todas las filas y todas las columnas excepto la última
df_features = df.iloc[:, :-1]

# Usa iloc para seleccionar solo la última columna
# Assuming your 'Clase' column contains 'Up' and 'Down'
df_labels = df.iloc[:, -1]

# Dividir los datos en conjuntos de entrenamiento, validación y prueba
train_data, train_labels = df_features[:train_size], df_labels[:train_size]
validation_data, validation_labels = df_features[train_size:train_size + validation_test_size], df_labels[train_size:train_size + validation_test_size]
test_data, test_labels = df_features[train_size + validation_test_size:], df_labels[train_size + validation_test_size:]

print(train_data.shape[0], train_labels.shape[0])
print(validation_data.shape[0], validation_labels.shape[0])
print(test_data.shape[0])

print(df.head())
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Definir el modelo LSTM
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(train_data.shape[1],)),  # Capa de entrada
    tf.keras.layers.Dense(128, activation='relu'),       # Capa densa con activación ReLU
    tf.keras.layers.Dense(64, activation='relu'),        # Otra capa densa con activación ReLU
    tf.keras.layers.Dense(1, activation='sigmoid')        # Capa de salida con activación sigmoide para clasificación binaria
])
# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
import numpy as np

# Convert DataFrame to NumPy array and set data type to float32
train_data_array = np.array(train_data, dtype='float32')
validation_data_array = np.array(validation_data, dtype='float32')
test_data_array = np.array(test_data, dtype='float32')

# Train the model
history = model.fit(train_data_array, train_labels, epochs=110, batch_size=16,
          validation_data=(validation_data_array, validation_labels))
```

```python
# Obtener las métricas de entrenamiento y validación del historial
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(train_accuracy) + 1)

# Graficar la precisión
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs, train_accuracy, 'b-', label='Training Accuracy')
plt.plot(epochs, val_accuracy, 'r-', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Graficar la pérdida
plt.subplot(1, 2, 2)
plt.plot(epochs, train_loss, 'b-', label='Training Loss')
plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Mostrar los gráficos
plt.show()
```

```python

# Evaluar el modelo en el conjunto de prueba
test_loss, test_accuracy = model.evaluate(tf.convert_to_tensor(test_data.values[:, :, tf.newaxis]), test_labels)
print(f'Test Accuracy: {test_accuracy}')

# Hacer predicciones en el conjunto de prueba
predictions = model.predict(tf.convert_to_tensor(test_data.values[:, :, tf.newaxis]))

# Puedes convertir las predicciones a clases ('Up' o 'Down') usando un umbral (por ejemplo, 0.5)
threshold = 0.5
predicted_classes = ['Up' if pred > threshold else 'Down' for pred in predictions]

# Imprimir las predicciones
print(predicted_classes)
```
